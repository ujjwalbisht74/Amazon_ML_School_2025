# Data Preprocessing Techniques

 - Rescale data / Normalization
 
 - It is used whenever your data consist of multiple features with different scale .
 - It will optimize algorithms for gradient descent computation / and also for distance based algorithms like Regression and Nueral Network.
 - It nullify the dominancy of high weight feature to prevent bias prediction
 
 - Normalizations Techniques : 
     - Min-Max-Scaling.
     - L1-L2-Normalization.

#### MIN-MAX SCALAR


```python
# import libraries
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from sklearn.preprocessing import MinMaxScaler
from sklearn.datasets import make_classification
```

```python
# create dataset
X, y = make_classification(n_samples=100, n_features=5, random_state=42)
```

```python
plt.figure(figsize=(12, 6))
plt.boxplot(X, labels=[f'Feature {i+1}' for i in range(X.shape[1])])
plt.title('Boxplot of Original Data')
plt.xlabel('Features')
plt.ylabel('Value Range')
plt.show()
```

```python
for i in range(1,6):
    print(f'Min of original Feature{i} :',np.min(X[i]))
    print(f'Max of original Feature{i} :', np.max(X[i]))
```

```python
# applying MinMax Scaling

# create object of minmaxscaler and fit it
scaler = MinMaxScaler().fit(X);

# tranform values
x_scaled = scaler.transform(X);
```

```python
plt.figure(figsize=(12, 6))
plt.boxplot(x_scaled, labels=[f'Feature {i+1}' for i in range(X.shape[1])])
plt.title('Boxplot of Original Data')
plt.xlabel('Features')
plt.ylabel('Value Range')
plt.show()
```

```python
for i in range(1,6):
    print(f'Min of scaled Feature{i} :',np.min(X[i]))
    print(f'Max of scaled Feature{i} :', np.max(X[i]))
```

### What we got ?
 - Values now are in same range (no weight are dominant).
 - It will make data follow normal distribution

## L1 - L2 Scaling 

- To understand this first you have to about norm.. 

- It describe size of vectors `set of numbers`.

- Suppose, we want to calculate size of vector , there is a way that we can compute it by calculating distance through origin. 

- Two types of  norms : 
    - `l1 norm / Manhattan Distance` : if we consider evry indivisual component of vector on itself and then used for calculation of distance form origin.
    - `l2 norm / Euclidian Distance` focused on 2D vector as a component for calculation of distance

### L1 Normalization

 - L1 normalization, also known as least absolute deviations (LAD), is a method of normalizing data based on the sum of absolute values of each element. It ensures that the sum of the absolute values of each element in a vector is equal to 1.
 
<centre>
     
<centre>
    
- L2 normalization, also known as least squares or Euclidean norm, scales data so that the sum of the squares of the values (magnitude) of each row (or sample) equals 1.
<centre>
    
<centre>

```python
from sklearn.preprocessing import Normalizer
```

```python
X, _ = make_classification(n_samples=100, n_features=2, n_redundant=0, n_clusters_per_class=1, random_state=42)


normalizer_l1 = Normalizer(norm='l1')
X_normalized_l1 = normalizer_l1.transform(X)

normalizer_l2 = Normalizer(norm='l2')
X_normalized_l2 = normalizer_l2.transform(X)


plt.figure(figsize=(12, 6))

plt.subplot(1, 3, 1)
plt.scatter(X[:, 0], X[:, 1], color='b')
plt.title('Original Data')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')


plt.subplot(1, 3, 2)
plt.scatter(X_normalized_l1[:, 0], X_normalized_l1[:, 1], color='g')
plt.title('L1 Normalized Data')
plt.xlabel('Feature 1 (Normalized)')
plt.ylabel('Feature 2 (Normalized)')


plt.subplot(1, 3, 3)
plt.scatter(X_normalized_l2[:, 0], X_normalized_l2[:, 1], color='r')
plt.title('L2 Normalized Data')
plt.xlabel('Feature 1 (Normalized)')
plt.ylabel('Feature 2 (Normalized)')

plt.tight_layout()
plt.show()
```

### Log Scaling 
    - power of 2 conversion of data samples
    - used for compression or explanding of ranges
    - it is particularly useful when dealing with data that spans several orders of magnitude, as it can make patterns more visible and data distributions more interpretable.

```python
# synthetic data 
np.random.seed(42)
data = np.random.exponential(scale=1.0, size=1000)

plt.hist(data, bins = 20)
plt.title('Original data')
plt.show()

new_data = np.log(data)

plt.hist(new_data, bins = 20)
plt.title('After Log Transformation')
plt.show()
```

### Binarizer

- Binary classification of data based on threshold . for a given threshold it will convert data into categories
    - values > threshold  : 1
    - values <= threshold : 0

```python
from sklearn.preprocessing import Binarizer

x = np.random.randn(100)

plt.hist(x)
plt.title('Original data')
plt.show()

x_reshaped = x.reshape(-1, 1)
scaler = Binarizer(threshold = 0.5)

new_x = scaler.transform(x_reshaped)

plt.hist(new_x)
plt.title('After Binarize')
plt.show();
```
