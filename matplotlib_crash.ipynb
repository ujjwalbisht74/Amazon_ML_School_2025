## Types Of Plotting
    - Basic Line Plot
    - Scatter Plot
    - Bar Plot
    - Histogram
    - Pie Chart
    - Subplots
    - Customization (Titles, Labels, Legends, Colors)
    - Annotations
    - Saving Figures
    

```python
# basic line plot 

# it contains x axis data and y axis data

x_axis = np.arange(12);
y_axis = np.linspace(1,100,12);


# adjust figure size
plt.figure(figsize= (10,7))
fig, ax  = plt.subplots()
ax.plot(x_axis,y_axis)


# how we can add titles and labels
plt.title('Basic Line plot')
plt.xlabel('x-axis')
plt.ylabel('y-axis')

# show the figure
plt.show()
```

### Scatter Plot


```python
# scatter function
plt.scatter(x_axis,y_axis, color = 'red')

# how to add label
plt.title('scatter-plot')

# how to add axis label
plt.xlabel('x-axis')
plt.ylabel('y-axis')
```

### Bar Plot

```python
# bar data consist of two things category and data(frequency/amount etc)

category = ['movies' , 'games' , 'travel' , 'racing']
likes = [12,20,100,50]

# bar function to plot bar 
plt.bar(category, likes, color = 'purple')

# setting titles and labels
plt.title('category-wise likes')
plt.xlabel('hobby')
plt.ylabel('no-of-likes')
```

### Histogram

```python
# simple data is required but you have to set the intervals 
import numpy as np
data  = np.random.randn(1000) # synthetic data

# how to plot hist
plt.hist(data , bins = 20 , color = 'blue')

# how to add title
plt.title('histogram')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
```

### Pie Charts

```python
# for pie charts we need  3 parameters (size,label, color)

size = [10,20,30,40]
labels = ['A', 'B', 'C', 'D']
colors = ['gold', 'yellowgreen', 'lightcoral', 'lightskyblue']

# plot piechart
plt.pie(size, labels = labels , colors = colors)

# how to add titles and labels
plt.title('Pie-Chart')
```

### SubPlots

```python
# you can create multiple plots on same axis

x = np.arange(5)
y = [1,2,3,7,11]
z = [3,5,7,9,13]

fig ,(ax1,ax2) = plt.subplots(1,2, figsize=(12, 5))

# setting up subplot 1
ax1.plot(x,y)
ax1.set_title('Subplot 1')
ax1.set_xlabel('x-axis')
ax1.set_ylabel('y-axis')

ax2.plot(x,z)
ax2.set_title('Subplot 2')
ax2.set_xlabel('x-axis')
ax2.set_ylabel('y-axis')

plt.show()
```

### Customizations

```python
x = [1, 2, 3, 4, 5]
y1 = [2, 3, 5, 7, 11]
y2 = [1, 4, 6, 8, 10]


plt.figure(figsize=(8, 5))

# new label
plt.plot(x, y1, label='Line 1', color='blue', linestyle='--', marker='o')
plt.plot(x, y2, label='Line 2', color='green', linestyle='-', marker='x')
```

### Annotate Plots

```python
x = [1, 2, 3, 4, 5]
y = [2, 3, 5, 7, 11]

plt.figure(figsize=(8, 5))
plt.plot(x, y, marker='o')

# Add annotation
for i, txt in enumerate(y):
    plt.annotate(txt, (x[i], y[i]))
    
plt.show()
```
