### Missing Data

```python
# missing values are shown by 'NaN'

data = pd.DataFrame(
    {
        'value1' : np.arange(13),
        'value2' : np.array([1,2,3,4,5,6,5,7,10,'NaN','NaN','NaN','NaN']),
        'value3' : pd.Timestamp('20210312')
    },
    index = pd.date_range("20210312",periods =13)
) 
data



# delete missing values entire row
data.dropna()
print(data)

# fill na values with arbitory value
print(df.fillna(4))


# to count Na values
print("Missing values : ",df.isnull().sum())
```

### Merge and Concat

```python
# you can combine data slices into single frame using concat

temp = pd.DataFrame(np.random.randn(10,4))

pieces = [df[:3], df[3:7], df[7:]]

pd.concat(pieces)
```

```python
# join as sql 

left = pd.DataFrame({"key": ["foo", "foo"], "lval": [1, 2]})

right = pd.DataFrame({"key": ["foo", "foo"], "rval": [4, 5]})

print(left,"\n\n",right)

pd.merge(left, right , on = "key" ) # combining two differnent table on common features 
```

### IMPORT CSV


```python
# load data using pandas 
data = pd.read_csv("diabetes.csv")

data.head()
```

```python
# some basic info about dataset 

print("No of Rows and coloumns :",data.shape) # row and columns 
print("Dimension of dataset :",data.ndim) # total dimension
print("Total records :",data.size)

print("\n")
print("Data Type of eash feature in dataset :\n",data.dtypes)
```

```python
# performing some descriptive statistics 

data.describe()

# count : tells about any missing values in each column.
# mean : most frequent values of the dataset (what most people have).
# std : how much data deviates from mean value.
# min : minimum values that lies in each variable.
# 25-percentile : how much data lies in half of first half of data 
```

```python
# check for class distribution
info = pd.DataFrame(data.groupby('Outcome').size())
print(info)
```

### Correlations

```python
# we are using Pearson correlation coffecient

# - it assumes data follow normal distribution
# - -1 show full negative correlation and +1 show positive correlation , 0 means no correlation at all

data.corr(method = 'pearson')
```

### Skew Analsysis for Univariate fields 

```python
data.skew()

# +ve : right skew
# -ve : left skew
# values closer to zero show less skew
```
